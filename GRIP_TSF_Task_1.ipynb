### Decision Trees

```python
# Importing libraries in Python
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the iris dataset
iris=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

### Now let us define the Decision Tree Algorithm

```python
# Defining the decision tree algorithm

from sklearn.tree import DecisionTreeClassifier, plot_tree
dtree=DecisionTreeClassifier()
dtree.fit(df,y)

print('Decision Tree Classifer Created')
```

### Let us visualize the Decision Tree to understand it better.



```python
# Import necessary libraries for graph viz

from six import StringIO
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

### Apply the decision tree classifier model to the data using all four parameters at once, but with a maximum tree depth of 3

```python
%matplotlib inline
model_all_params_max_depth_3 = DecisionTreeClassifier(max_depth = 3).fit(df, y)
# Prepare a plot figure with set size.
plt.figure(figsize = (12,8))
# Plot the decision tree.
plot_tree(model_all_params_max_depth_3,
          rounded = True,
          filled = True                )
# Display the tree plot figure.
plt.show()
```

### Apply the model to the data as before, but with a minimum impurity decrease of 0.01

```python
model_all_params_min_imp_dec_001 = DecisionTreeClassifier(min_impurity_decrease = 0.01).fit(df, y)
# Prepare a plot figure with set size.
plt.figure(figsize = (16,10))
# Plot the decision tree.
plot_tree(model_all_params_min_imp_dec_001,
          rounded = True,
          filled = True                )
# Display the tree plot figure.
plt.show()
```

### Apply the decision tree classifier model to the data using all four parameters at once.

```python
model_all_params = DecisionTreeClassifier().fit(df, y)
# Prepare a plot figure with set size.
plt.figure(figsize = (20,10))
# Plot the decision tree, showing the decisive values and the improvements in Gini impurity along the way.
plot_tree(model_all_params, 
          filled=True      )
# Display the tree plot figure.
plt.show()
```
